### Completando con datos de otras competiciones

Completamos el dfKaggle con las competitiones que no tienen premio economico.

```python
ListTotalCompetitions = list(LeaderBoardPublicReady.competitionId)
```

```python
competitionsNoReward = list(set(ListTotalCompetitions ) - set(ListCompetitionReward)) 
```

```python
dfCompetitionNoReward = LeaderBoardPublicReady[LeaderBoardPublicReady["competitionId"].isin(competitionsNoReward)]
```

```python
dfCompetitionNoReward.dtypes
```

```python
competitionsReady = competitionsReady[competitionsReady["competitionId"] != 'tart']
competitionsReady["competitionId"] = competitionsReady["competitionId"].astype(float).astype(int)
```

```python
dfCompetitionNoReward_1 = pd.merge(dfCompetitionNoReward, competitionsReady, how = "inner", on = ["competitionId"])
dfCompetitionNoReward_1.head()
```

```python
dfCompetitionNoReward_2 = dfCompetitionNoReward_1[dfCompetitionNoReward_1["totalTeams"] == 0]
```

```python
len(dfCompetitionNoReward_2.competitionId.unique())
```

```python
dfCompetitionNoReward_2 = dfCompetitionNoReward_1[dfCompetitionNoReward_1["totalTeams"] != 0]
```

```python
dfCompetitionNoReward_2.drop_duplicates()
```

```python
dfKaggle.columns
```

```python
dfCompetitionNoReward_2["change"] = None
dfCompetitionNoReward_2["deadline"] = None
dfCompetitionNoReward_2['rank'] = None
dfCompetitionNoReward_2['score'] = None
dfCompetitionNoReward_2['teamMembers'] = None
dfCompetitionNoReward_2['rewardQuantity'] = None
dfCompetitionNoReward_2['evaluationMetric'] = None
```

```python
DtentriesNoreward= dfCompetitionNoReward_2.groupby('competitionId').agg({"competitionId":"count"})
DtentriesNoreward = DtentriesNoreward.rename(columns = {"competitionId" : "entries"})
DtentriesNoreward = DtentriesNoreward.reset_index()
DtentriesNoreward.head()
```

```python
dfCompetitionNoReward_3 = pd.merge(dfCompetitionNoReward_2, DtentriesNoreward, how = "inner", on = ["competitionId"])
dfCompetitionNoReward_3.head()
```

```python
dfKaggle.head()
```

```python
dfCompetitionNoReward_3.head()
```

```python
dfKaggle = dfKaggle.where(dfKaggle.notnull(), dfKaggle.astype(str))
dfCompetitionNoReward_3 = dfCompetitionNoReward_3.where(dfCompetitionNoReward_3.notnull(), dfCompetitionNoReward_3.astype(str))

dfKaggleCompleted = pd.concat([dfKaggle,dfCompetitionNoReward_3])
```
