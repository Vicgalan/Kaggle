```python
# Importacion de librerias

import pandas as pd

from os import listdir
from os.path import isfile, join
```

```python
# Posicionarse en un ruta donde sea se pueda leer el fichero comodamente

%cd C:\Users\Usuario\Desktop\Kaggle\Data\dataFramesParsed\Private_LeaderBoard_Parsed

# Lectura de los ficheros privateParse y concatenacion de los mismos.

leaderBoard_private_files = [f for f in listdir("Data_Private_Parsed") 
             if isfile(join("Data_Private_Parsed", f))]
```

```python
%cd C:\Users\Usuario\Desktop\Kaggle\Data\dataFramesParsed\Private_LeaderBoard_Parsed\Data_Private_Parsed

frames = []
for csv in leaderBoard_private_files:
    csv_name = csv
    df = pd.read_csv(csv_name, index_col=0)
    frames.append(df)
    
pd.set_option('display.max_colwidth', -1)
Dt_leaderBoard_private = pd.concat(frames)
```

Los principales cambios que se realizarán en este dataFrame son:
    - Renombrar las columnas con los nombres normalizados
    - Convertir la columna "entries" a Int
    - Convertir la columna "rank_leaderboard" a Int
    - Obtener una nueva columna "change", a partir de "position_change" y "position_change_reference"

#### Renombrar las columnas con los nombres normalizados

```python
Dt_leaderBoard_private = Dt_leaderBoard_private.rename(columns={'Evaluation_Score':'evaluationMetric',
                          'last_entry':'date',
                          'Evaluation_Score':'evaluationMetric',
                          'number_entries':'entries',
                          'name_team':'teamName',
                          'score_leaderboard':'score',                                    
                          'rank_leaderboard':'rank',
                          'team_members':'teamMembers',
                          'title':'competitionTitle',                              
                                 })
```

#### Convertir la columna "entries" a Int

```python
Dt_leaderBoard_private['entries'] = Dt_leaderBoard_private['entries'].astype(float).astype(int)
```

#### Convertir la columna "rank" a Int

```python
Dt_leaderBoard_private['rank'] = Dt_leaderBoard_private['rank'].astype(float).astype(int)
```

#### Obtener una nueva columna "change", a partir de "position_change" y "position_change_reference"

```python
Dt_leaderBoard_private['position_change'] = Dt_leaderBoard_private['position_change'].replace(['—'], 0)
Dt_leaderBoard_private['position_change'] = Dt_leaderBoard_private['position_change'].astype(float).astype(int)
```

```python
def set_change(row):
    if row["position_change_reference"] == "=":
        return 0
    elif row["position_change_reference"] == "-":
        return row["position_change"]*(-1)
    else:
        return row["position_change"]

Dt_leaderBoard_private = Dt_leaderBoard_private.assign(change=Dt_leaderBoard_private.apply(set_change, axis=1))

Dt_leaderBoard_private
```

```python
Dt_leaderBoard_private = Dt_leaderBoard_private.drop(columns=["position_change","position_change_reference"])
```
